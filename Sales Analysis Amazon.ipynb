```python
#importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/aksha/Desktop/Unified Mentor Internship Projects/Amazon Sales Data\Amazon Sales data.csv") #data importing
```

```python
data.head() #Checking the data (first 5 rows)
```

```python
data.tail() #(last 5 rows)
```

```python
# Data Cleaning Process

missing_values = data.isnull().sum()        #This function checks if the missing values are present inside the data or not
print("Missing_values:\n",missing_values)   #print missing values
```

```python
data["Order Date"] = pd.to_datetime(data["Order Date"]) #This will convert Order Date yo DateTime
```

```python
data["Year"] = data["Order Date"].dt.year
data["Month"] = data["Order Date"].dt.month
data.head()                                     #To extract yera and month from Order DateTime
```

```python
regions = data["Region"].nunique()   #To Calculate the number of regions
print("Number of Regions:",regions)  #To print
```

```python
country = data["Country"].nunique()   #To Calculate the number of countries
print("Number of Countries:",country) #To print
```

```python
item_type = data["Item Type"].nunique() #To calculate the item types
print("Number of Item Types:",item_type) #To print
```

```python
unit_sold = data["Units Sold"].sum() #To calculate the total units sold
print("Total Unit Sold:",unit_sold)  #To print
```

```python
unit_cost = data["Unit Cost"].sum()  #To calculate Total Unit cost
print("Total Unit cost:",unit_cost)  #To print
```

```python
total_cost = data["Total Cost"].sum()  #To calculate Total  cost
print("Total cost:",unit_cost)  #To print
```

```python
total_revenue = data["Total Revenue"].sum()  #To calculate Total Revenue
print("Total Revenue:",total_revenue)  #To print
```

```python
total_profit = data["Total Profit"].sum()  #To calculate Total profit
print("Total Profit:",total_profit)  #To print
```

```python
data.groupby(['Region','Sales Channel'])['Total Profit'].sum()   #This groupby function to bring all the calculations
```

```python
year_sales = data.groupby('Year')['Total Revenue'].mean()
plt.figure(figsize=(8,3))
sns.barplot(x=year_sales.index,y=year_sales.values,)
plt.title('Average Sales By Year')
plt.xlabel('Year')
plt.ylabel('Total Revenue')    #This is to Calculate Year Wise Sales 
```

```python
plt.figure(figsize=(6,6))
region_TotalRevenue = data.groupby('Region') ['Total Profit'].mean()
plt.pie(region_TotalRevenue,startangle=90,labels=region_TotalRevenue.index,autopct='%1.1f%%')
plt.title('Average Profit in Region wise')  #Pie chart Total profit in region wise
```

```python
TotalRevenue_ItemType = data.groupby('Item Type')['Total Revenue'].sum()  #To Grouping Total Revenue by Item Type
```

```python
plt.figure(figsize=(8,3))
TotalRevenue_ItemType.plot(kind='bar')
plt.xlabel('Item Type')
plt.ylabel('Total Revenue')
plt.title('Average Revenue by Product type')
plt.grid(axis='y')                                #Barchart for total revenue for item Type
```

```python
TotalRevenue_SalesChannel = data.groupby('Sales Channel')['Total Revenue'].mean() #group total revenue by sales channel
```

```python
plt.figure(figsize=(6,6))
plt.tight_layout()
TotalRevenue_SalesChannel.plot(kind='pie',autopct='%1.1f%%' ,startangle=90)
plt.title('Total Revenue by Sales Channel')   #bar chart for Total Revenue by Item type 
```

```python
Region_UnitSold = data.groupby ('Region' )['Units Sold']. sum()
plt.figure(figsize= (6,6))
Region_UnitSold.plot(kind='pie',labels=Region_UnitSold.index,autopct='%1.1f%%', startangle=90)  #To create a pie chart for a donut chart

                      
                     
cntr_circle = plt.Circle((0,0), (0.70),fc='white')  #To Draw a circle at the centre of the pie Chart 
fig = plt.gcf()
fig.gca().add_artist(cntr_circle)


plt.title('Units Sold by Regions')    #To equal aspect ratio ensures that pie is drawn as a circle
plt.axis ('equal')
```

```python
YearMonth_UnitsSold =data.groupby(['Year', 'Month']) ['Units Sold'].sum() #To Group by Units Sold by year and Month
```

```python
plt.figure(figsize= (6,6))
YearMonth_UnitsSold.plot(kind='bar')
plt.xlabel('Year and Month')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.grid(axis='y')
```

```python
TotalCost_SalesChannel = data.groupby('Sales Channel')['Total Cost'].sum()

plt.figure(figsize=(6,6))
TotalCost_SalesChannel.plot(kind='pie',autopct='%1.1f%%',startangle=90)
plt.title('Total Cost by Sales Channel')
plt.tight_layout()                 #To make Bar chart for total Cost by Sales Channel
```
